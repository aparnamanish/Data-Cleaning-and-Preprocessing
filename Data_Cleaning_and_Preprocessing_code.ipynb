```python
import pandas as pd
df = pd.read_csv('Medical Appointmen t No Shows dataset.csv')
df.head()
```

```python
df.shape
```

```python
#To identify the column names
for col in df.columns:
    print(repr(col))
```

```python
# To see the column names and data types
df.info()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Summary of numeric columns
df.describe()
```

```python
# Count duplicate rows
duplicates = df.duplicated().sum()
print("Duplicate rows:", duplicates)

# Remove duplicates if any
df = df.drop_duplicates()
```

```python
# Standardize text/categorical values without stripping or changing other letters
df['Gender'] = df['Gender'].replace({'M':'Male', 'F':'Female'})
df['No-show'] = df['No-show'].replace({'yes':'Yes', 'no':'No'})
df.head()
```

```python
# Clean and uniform column names(in lower case and without space)
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head()
```

```python
# Convert numeric columns to integer type
df['age'] = df['age'].astype(int)
df['scholarship'] = df['scholarship'].astype(int)
df['hipertension'] = df['hipertension'].astype(int)
df['diabetes'] = df['diabetes'].astype(int)
df['alcoholism'] = df['alcoholism'].astype(int)
df['handcap'] = df['handcap'].astype(int)
df['sms_received'] = df['sms_received'].astype(int)
```

```python
print("Cleaned dataset summary:")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
print("Any missing values:", df.isnull().values.any())
print("Duplicates removed:", duplicates)
```

```python
df.to_csv('cleaned_data.csv', index=False)
print("✅ Cleaned dataset saved as cleaned_data.csv")
```
